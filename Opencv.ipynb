```python
#Parameters
width = 640
height = 480
min_prob_val = 0.65 #

# Loading the model
json_file = open("model-bw.json", "r")
model_json = json_file.read()
json_file.close()

# Explicitly import the necessary classes
from tensorflow.keras.layers import InputLayer, Conv2D  # Import necessary layers

# Use keras.models.model_from_json to load the model, passing custom_objects if needed
loaded_model = model_from_json(model_json, custom_objects={'Sequential': tf.keras.models.Sequential,
                                                           'InputLayer': tf.keras.layers.InputLayer,
                                                           'Conv2D': tf.keras.layers.Conv2D
                                                           # Add other custom layers if used in your model
                                                          })
# load weights
loaded_model.load_weights("model-bw.h5")
print("Loaded model from disk")

#Preprossessing function
def prepos(img):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.equalizeHist(img)
    img = img/255
    return img
```

```python
classes = {
    0:'Apple',
    1:'Banana',
    2:'Lemon',
    3:'Stawberry',} ##['Apple', 'Banana', 'Lemon', 'Stawberry'] #['0', '1', '2', '3']
```

```python
import cv2

cap = cv2.VideoCapture(0)  # Try default camera first

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error opening default camera (index 0). Trying alternative cameras...")
    for i in range(10):  # Check camera indices from 1 to 9
        cap = cv2.VideoCapture(i)
        if cap.isOpened():
            print(f"Found camera at index {i}")
            break
    else:  # If no camera found
        print("Error opening camera. Please check camera connection and configuration.")
        raise RuntimeError("Could not open camera.")

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 100)  # Set width
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 100)  # Set height
```

```python
while True:
    success, imgOriginal = cap.read()

    # Check if the frame was read successfully
    if not success:
        print("Error reading frame from camera.")
        break  # Exit the loop if frame reading fails

    img = np.asarray(imgOriginal, dtype=np.uint8)  # Convert to uint8 explicitly
    img = cv2.resize(img, (32, 32))
    img = prepos(img)
    cv2.imshow("Processed Image", img)
    img = img.reshape(1, 32, 32, 1)
    
    # Predict
    classIndex = np.argmax(loaded_model.predict([img]), axis=-1)[0]
    sign = classes[classIndex]
    predictions = loaded_model.predict(img)
    probVal = np.amax(predictions)
    
    print(sign, probVal)

    if probVal > min_prob_val:
        cv2.putText(imgOriginal, str(sign) + "   " + str(probVal),
                    (50, 50), cv2.FONT_HERSHEY_COMPLEX,
                    1, (0, 0, 255), 1)

    cv2.imshow("Original Image", imgOriginal)

    interrupt = cv2.waitKey(10)
    if interrupt & 0xFF == 27:  # ESC key
        break

cap.release()
cv2.destroyAllWindows()
```
